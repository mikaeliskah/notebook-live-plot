```python
# matplotlib magic is necessary to display plots and to make them zoomable etc
# otherwise they would be in-line figures
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


# define sine function (less type-work)
def sin(x):
    # returns the sine function (for less typing)
    return np.sin(x)
    

# fft function
def fft(x):
    return np.fft.fftshift(np.fft.fft(x))    


# animation
def animate(f):
    '''this function will be animated
    a signal and its corresponding fft 
    will be calculated and displayed'''
    
    s = sin(2 * pi * f * t)
    sp = fft(s)
    ax1.clear()
    ax1.plot(t, s, label='sine')
    ax1.legend(loc='upper right')
    ax1.set_ylabel('A')
    ax1.set_xlabel('time [s]')
    ax1.set_xlim(0,0.1)
    
    ax2.clear()
    ax2.plot(sp, label='fft')
    ax2.legend(loc='upper right')
    

# initialize a figure
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# constants and variables:
pi = np.pi # pi
N = 1000   # datalength
T = 1      # Time-Window
dt = T / N # time-interval
f_max = .5 * (N / T)
t = np.linspace(0,T,N)   
f_steps = 100 # number of frequencies to cycle through 
f_list = np.linspace(-f_max,f_max, f_step) # frquency-list

# animate
ani = FuncAnimation(fig, animate, f_list, interval=100)

# show les plottes
plt.show()
# prevent subplots from overlapping
plt.tight_layout()
```
